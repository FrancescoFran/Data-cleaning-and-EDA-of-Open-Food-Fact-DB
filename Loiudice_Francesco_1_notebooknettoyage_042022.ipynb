## Réduction des variables avec l'évaluation du taux de remplissage

```python
#fonction pour calculer le taux de remplissage des variables
def taux_remplissage(dataset,variable):
    taux_remplissage=(1-(dataset[variable].isna().sum()/len(dataset[variable])))*100
    return taux_remplissage

taux_rempl=[]
for col in data.columns:
    taux_rempl.append(taux_remplissage(data,col))

#création dataframe taux de remplissage du jeu de données ordonnée de façon déscendante
Taux_df=pd.DataFrame({'Variable':data.columns,'Taux de remplissage':taux_rempl},columns=['Variable','Taux de remplissage']).sort_values('Taux de remplissage',ascending=False)
```

```python
#définition seuil du taux de remplissage pour suppression variable après évaluation du minimum entre taux nutrition grade et taux_pnns_groups_1 
print('Taux remplissage variable nutrition_grade_fr_100g: {:.1f}'.format(float(taux_remplissage(data,'nutrition_grade_fr'))))
print('Taux remplissage variable pnns_groups_1: {:.1f}'.format(float(taux_remplissage(data,'pnns_groups_1'))))
Taux_min=min(float(taux_remplissage(data,'nutrition_grade_fr')),float(taux_remplissage(data,'pnns_groups_1')))
#on fixe le seuil de remplissage à 28%
sup_threshold=28
print('Seuil taux de remplissage: {:.1f}'.format(sup_threshold))
```

### Correction des valeurs aberrantes de la variable 'pnns_groups_1'

```python
def nettoyage_var_qualit(variable,seuil):
    #analyse des individus uniques et replacement des valeurs similaires
    Var=variable
    Var_list=data[Var]
    Var_list.dropna(inplace=True)
    Var_list=Var_list.unique()
    print('*'*100)
    l=len(Var_list)
    print('Nombre valeurs uniques en ',Var,'variable: ',l)
    print('\nValeurs uniques en ',Var,'variable:\n',Var_list)
    element1_list=[] #liste des individus uniques analysés
    element2_list=[] #liste des individus uniqués à analysér
    cnt1=0
    cnt2=0
    i=0
    Total_replaced_elem=0
    
    for element1 in Var_list:
        i+=1
        if element1 not in element1_list:      
            element1_list.append(element1)
            Var_list=np.delete(Var_list,0)
            for element2 in Var_list:
                Score=fuzz.ratio(element1,element2)
                if (Score>seuil) and np.any(data[Var]==element1):
                    print('-'*100)
                    print(element1)
                    #décompte des element1
                    cnt1=data[Var].value_counts()[data[Var].value_counts().index==element1].values
                    #décompte des element2
                    cnt2=data[Var].value_counts()[data[Var].value_counts().index==element2].values
                    if cnt1>cnt2: #on choisit la valeur plus fréquente
                        Remain_element=element1
                        Remain_element_cnt=cnt1
                        Replaced_element=element2
                        Replaced_element_cnt=cnt2
                        data[Var].replace(Replaced_element,Remain_element,inplace=True)
                    else:
                        Remain_element=element2
                        Remain_element_cnt=cnt2
                        Replaced_element=element1
                        Replaced_element_cnt=cnt1
                        data[Var].replace(Replaced_element,Remain_element,inplace=True)
                    print('Step ',i,'/',l,' : ',Replaced_element,'(counted ',Replaced_element_cnt,'times in', Var,') replaced with',Remain_element,'(counted ',Remain_element_cnt,'times in ',Var,' )')
                    print("Score de similarité:",Score,'%')
                    Total_replaced_elem+=Replaced_element_cnt

    print('\nNombre totale valeurs corrigées dans ',Var,': ',Total_replaced_elem,'\n')
    return()

data['pnns_groups_1']=np.where(data['pnns_groups_1']=="unknown",np.nan,data['pnns_groups_1']) #on remplace la catégorie 'unknown' avec NaN
nettoyage_var_qualit('pnns_groups_1',80)
```

## Remplissage avec moyenne ou médiane

```python
#histogrammes et Kolmogorov test
import scipy.stats as stats

variables=['energy_100g',
           'fat_100g',
           'saturated-fat_100g',
           'trans-fat_100g',
           'cholesterol_100g',#
           'carbohydrates_100g',
           'sugars_100g',
           'fiber_100g',
           'proteins_100g',
           'salt_100g',
           'sodium_100g',
           'vitamin-a_100g',
           'vitamin-c_100g',
           'calcium_100g',
           'iron_100g'         
          ]

fig=plt.figure(figsize=[15,70])
n=1
for var in variables:
    k=round(1+np.log2(data[var].shape[0])) #nombre de bins dans l'histogramme
    fig.add_subplot(16,2,n)
    ax=sns.histplot(data[var], element="bars",stat="density",bins=k,kde=False,label=var)
    #on calcule la fonction de distribution  
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 10000)
    filtered_data = data[var][data[var].notnull()]
    mu, std = stats.norm.fit(filtered_data)
    p = stats.norm.pdf(x, mu, std)
    ax.plot(x, p, 'r', lw=2, label='pdf')
    kstest = stats.kstest(data[var].notnull(),'norm')
    ax.set_title("Distribution vs loi normale de {} ".format(var))
    ax.text(1, 1, 'Test Kolmogorov-Smirnov \n Pvalue: {:.2} \n Stat: {:.2}'.format(kstest.pvalue, kstest.statistic),
            horizontalalignment='right',
            verticalalignment='top',
            style='italic', transform=ax.transAxes, fontsize = 12,
            bbox={'facecolor':'#00afe6', 'alpha':0.5, 'pad':0})
    n=n+1
plt.tight_layout()
plt.show()
```

```python
pd.options.mode.chained_assignment = None

data_subset=data[['vitamin-a_100g','vitamin-c_100g','calcium_100g','iron_100g','trans-fat_100g','cholesterol_100g']]
for col in data_subset:
    df=data[['pnns_groups_1',col]]
    #pour les valeurs de la variable ou le pnns groupe est présent, on remplit les valeurs manquantes avec la médiane du groupe
    df.loc[df['pnns_groups_1'].notnull(),col]=df[col].fillna(df.groupby('pnns_groups_1')[col].transform('median'))
    #pour les valeurs de la variable ou le pnns groupe est présent, remplissage avec médiane des valeurs avec NaN dans pnns_groupe_1 
    m=np.nanmedian(df[col])
    data_subset[col]=df[col].fillna(m)
```

```python
from sklearn.impute import KNNImputer

nutrition_grade_num=data['nutrition_grade_fr'].map({'a':0,'b':1,'c':2,'d':3,'e':4})
```

```python
data_quantit=data.select_dtypes(exclude='object')
data_quantit['nutrition-score-fr_100g']=nutrition_grade_num
#remplissage du nutrigrade avec KNN avec les variables nutritionnelles
data_quantit=data_quantit.rename({'nutrition-score-fr_100g':'nutrition_grade_fr'},axis=1)
```

```python
imputer = KNNImputer(n_neighbors=2)
imputed =imputer.fit_transform(data_quantit)
imputed_data_subset = pd.DataFrame(imputed, columns=data_quantit.columns)
imputed_data_subset['nutrition_grade_fr']=imputed_data_subset['nutrition_grade_fr'].astype(int)
imputed_data_subset['nutrition_grade_fr'].replace({0:'a',1:'b',2:'c',3:'d',4:'e'},inplace=True)
```

```python
data['nutrition_grade_fr']=imputed_data_subset['nutrition_grade_fr'].values
```
